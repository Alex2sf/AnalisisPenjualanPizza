```python
#1.For each order, which is their total price?

order_price = df.groupby(['order_id']).agg({'total_price':'sum'})
sns.displot(order_price, bins=20, palette = "Dark2", legend = False, kde = True)
plt.xlabel('Price of orders')
plt.ylabel('Frequency')
plt.title('Distribution of order prices');    
```

```python
#2. Which size of pizza is ordered the most?

total = len(df) # total number of orders
L = len(df[df.pizza_size == 'L']) / total # % of Large pizzas to total no. of orders
M = len(df[df.pizza_size == 'M']) / total # % of Medium pizzas to total no. of orders
S = len(df[df.pizza_size == 'S']) / total # % of Small pizzas to total no. of orders
XL = len(df[df.pizza_size == 'XL']) / total # % of Extra Large pizzas to total no. of orders
XXL = len(df[df.pizza_size == 'XXL']) / total # % of Extra Extra Large pizzas to total no. of orders

# Plot the pie chart
figure, ax = plt.subplots()
patches, texts, pcts = ax.pie([L,M,S,XL+XXL], 
                              labels = ["L","M","S","XL+XXL"],
                              autopct = '%.1f%%',
                              wedgeprops={'linewidth':1.0, 'edgecolor': 'white'},
                              textprops = {'size': 'x-large'},
                              shadow = False)

# Title of pie chart
ax.set_title('Pizza sizes by orders (%)', fontsize = 15)

# Annotation
for i, patch in enumerate(patches):
    texts[i].set_color(patch.get_facecolor())
plt.setp(pcts, color='white')
plt.setp(texts, fontweight = 600)

# Show the pie chart
plt.tight_layout();
```

```python
# Show the number of orders for each category of pizza
categories = df.pizza_category.value_counts()

# Find the index of the highest value (highest number of orders)
highest_index = categories.idxmax()

# Set the color of the highest bar to green, and others to blue color
colors = ['green' if i == highest_index else 'blue' for i in categories.index]

# Plot the bar chart
plots = categories.plot.barh(figsize = (10,5), color = colors)
plt.ylabel('Pizza category')
plt.xlabel('Number of orders')
plt.title('Number of orders by pizza categories', fontsize = 15)

# Total number of orders
total = len(df)

# Iterating over the bars one-by-one
for bar in plots.patches:
   
  # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # y-coordinate: bar.get_y() + bar.get_width() / 2
  # x-coordinate: bar.get_width()
  # ha and va stand for the horizontal and vertical alignment
    plots.annotate(format(bar.get_width() / total, '.0%'),
                   (bar.get_width(), bar.get_y() + bar.get_height() / 2),
                    ha='center', va='center',
                   size=13, xytext=(-25, 0),
                   textcoords='offset points', color = 'white', fontweight = 600)
    
# Hide the right and top spines
plots.spines[['right', 'top']].set_visible(False);
```

```python
#Kombinasi Pizza yang Paling Sering Dipesan: Menganalisis kombinasi pizza yang sering dipesan bersama untuk mengetahui preferensi pelanggan.

import pandas as pd
from itertools import combinations
from collections import Counter


# Menggabungkan pizza_name berdasarkan order_id
orders = df.groupby('order_id')['pizza_name'].apply(list)

# Menghitung kombinasi pizza yang dipesan bersama
combination_counts = Counter()
for order in orders:
    if len(order) > 1:  # Hanya mempertimbangkan order dengan lebih dari satu pizza
        combs = combinations(order, 2)
        combination_counts.update(combs)

# Menampilkan 10 kombinasi pizza yang paling sering dipesan bersama
most_common_combinations = combination_counts.most_common(10)

# Membuat DataFrame untuk visualisasi
combos_df = pd.DataFrame(most_common_combinations, columns=['Combination', 'Count'])

# Visualisasi dengan matplotlib dan seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y='Combination', data=combos_df, palette='viridis')
plt.xlabel('Count')
plt.ylabel('Pizza Combination')
plt.title('10 Most Common Pizza Combinations Ordered Together')
plt.show()
```

```python
#5. How many pizzas are we making during peak periods?
# Add two new columns to df: day_of_week, day_name
day_of_week = []
day_name = []

for i in range(len(df)):
    # 0 = Mon, 1 = Tue, 2 = Wed, ..., 6 = Sun
    day_of_week.append(df.order_date[i].dayofweek)
    
    # Monday, Tuesday, ..., Sunday
    day_name.append(df.order_date[i].day_name()) 

df['day_of_week'] = day_of_week
df['day_name'] = day_name

# Count number of orders by each day
busiest_day_df = df.groupby(['day_of_week','day_name']).count().sort_index(ascending=True).reset_index().set_index('day_of_week')

# Find the index of the highest value (highest number of orders by day)
highest_index = busiest_day_df['total_price'].idxmax()

# Set the color of the highest bar to green, and others to blue color
colors = ['green' if i == highest_index else 'blue' for i in busiest_day_df.index]

# Plot the bar chart
ax = busiest_day_df.plot.bar(x = 'day_name', y = 'total_price', figsize = (15,7), legend=False, color = colors)
plt.ylabel("Total revenue by day")
plt.xlabel("Day of week")
plt.title("Which day tend to be the most busiest?")
ax.set_xticklabels(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
                  rotation = 0);

# Iterating over the bars one-by-one
for i, bar in enumerate(ax.patches):
    ax.annotate(format(bar.get_height(), '.0f'),
                (bar.get_x() + bar.get_width() / 2,
                bar.get_height()), ha='center', va='center',
                size=10, xytext=(0, 8),
                textcoords='offset points', color = colors[i], fontweight = 600)
    
# Hide the right and top spines
ax.spines[['right', 'top']].set_visible(False);

```

```python
# Now we look to investigate Friday times.
friday_df = df[df.day_name == 'Friday']
friday_df
```

```python
# Convert order_time into datetime data type
friday_df.order_dummydatetime = pd.to_datetime(friday_df.order_time, format = '%H:%M:%S')

# Define the bins
bins = np.arange(0,26,2)
print(bins)

# Add labels (12 bins, each bin represents 2 hours window)
labels = ['00:00-01:59', '02:00-03:59', '04:00-05:59', '06:00-07:59','08:00-09:59','10:00-11:59',
         '12:00-13:59', '14:00-15:59', '16:00-17:59','18:00-19:59','20:00-21:59','22:00-23:59']

# Add the bins to friday_df
friday_df['time_bin'] = pd.cut(friday_df.order_dummydatetime.dt.hour, bins, labels=labels, right=False)
```

```python
friday_df
```

```python
# The number of orders during each time bin
orders_per_bin = pd.DataFrame(friday_df.time_bin.value_counts())
orders_per_bin.reset_index(inplace=True)
orders_per_bin.columns = ['time_bin','count']
orders_per_bin['time_bin'] = pd.Categorical(orders_per_bin['time_bin'], categories=labels, ordered=True)
sorted_df = orders_per_bin.sort_values('time_bin', ascending=False).reset_index(drop=True)

# Display the DataFrame (sorted by time bins from 12am to 11:59pm)
sorted_df.set_index("time_bin", inplace=True)
sorted_df
```

```python
sorted_df.plot.barh(legend = False, figsize = (15,7), color = 'blue')
plt.xlabel('Number of orders per time bin')
plt.title("The peak times of the pizza restaurant on Fridays (number of orders from each time bin)");
```

```python
# Revenue earned from each time bin
revenue_friday_df = pd.DataFrame(friday_df.groupby('time_bin').total_price.sum())
revenue_friday_df.reset_index(inplace=True)
revenue_friday_df = revenue_friday_df.sort_values("time_bin",ascending=False)
revenue_friday_df.set_index("time_bin").plot.barh(figsize = (15,7), legend = False, color = 'blue')
plt.xlabel("Total revenue earned from each time bin")
plt.ylabel("Time bin")
plt.title("The peak times of the restaurant on Fridays (revenue earned from each time bin)");
```

```python
#12. Which month was revenue earned the highest?
# Extract the month and month_name from order_date
month = []
month_name = []
for i in df.order_date:
    # 1 = January, 2 = February, ..., 12 = December
    month.append(i.month)
    
    # January, February, ..., December
    month_name.append(i.month_name())
    
df['month'] = month
df['month_name'] = month_name
total_revenue = df.total_price.sum() # total revenue = sum of prices of all orders

# Create a colormap from red to yellow
cmap = mcolors.LinearSegmentedColormap.from_list("", ["green", "yellow", "red"])

# Find the total revenue by each month using groupby
revenue_per_month_df = df.groupby(['month', 'month_name']).agg({'total_price': 'sum'})

# Rank the monthly total revenue from highest to lowest
revenue_per_month_df['rank'] = revenue_per_month_df['total_price'].rank(method="dense", ascending=False)

# Sort from the lowest revenue to the highest revenue
revenue_per_month_df = revenue_per_month_df.sort_values('rank', ascending=False).reset_index()
revenue_per_month_df
```

```python
# Plot the bar chart
fig, ax = plt.subplots(figsize = (15,7))
ax.barh(revenue_per_month_df.month_name, revenue_per_month_df['total_price'],
        color=cmap(revenue_per_month_df['rank']/12))
ax.set_ylabel('Month')
ax.set_xlabel('Revenue earned by month')
ax.set_title('Monthly revenue in 2015 (from highest to lowest)')

# Annotation: iterating over the bars one-by-one
for i, bar in enumerate(ax.patches):
    ax.annotate(f"USD {revenue_per_month_df['total_price'][i]:.2f} ({bar.get_width() / total_revenue:.2%})",
                   (bar.get_width(), bar.get_y() + bar.get_height() / 2),
                    ha='center', va='center',
                   size=14, xytext=(-100, 0),
                   textcoords='offset points', color = 'navy', fontweight = 400)
    
# Hide the right and top spines
ax.spines[['right', 'top']].set_visible(False);
```

```python
#Total Penjualan per Hari/Bulan/Tahun: Menghitung total penjualan harian, bulanan, atau tahunan untuk mengidentifikasi tren penjualan.
# Pastikan kolom 'order_date' dalam format datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Total penjualan per hari
daily_sales = df.groupby(df['order_date'].dt.date)['total_price'].sum()

# Total penjualan per bulan
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['total_price'].sum()

# Total penjualan per tahun
yearly_sales = df.groupby(df['order_date'].dt.to_period('Y'))['total_price'].sum()

# Menampilkan hasil
print("Total Penjualan per Hari:")
print(daily_sales.head())

print("\nTotal Penjualan per Bulan:")
print(monthly_sales.head())

print("\nTotal Penjualan per Tahun:")
print(yearly_sales.head())

# Visualisasi dengan matplotlib
plt.figure(figsize=(15, 5))

# Visualisasi penjualan harian
plt.subplot(1, 3, 1)
daily_sales.plot(kind='line')
plt.title('Total Penjualan per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')

# Visualisasi penjualan bulanan
plt.subplot(1, 3, 2)
monthly_sales.plot(kind='line')
plt.title('Total Penjualan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Penjualan')

# Visualisasi penjualan tahunan
plt.subplot(1, 3, 3)
yearly_sales.plot(kind='line')
plt.title('Total Penjualan per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Penjualan')

plt.tight_layout()
plt.show()
```

```python
#Penjualan Berdasarkan Kategori Pizza: Mengetahui kategori pizza mana yang paling laris.
# Menghitung total penjualan per kategori pizza
category_sales = df.groupby('pizza_category')['total_price'].sum().sort_values(ascending=False)

# Menampilkan hasil
print("Total Penjualan Berdasarkan Kategori Pizza:")
print(category_sales)

# Visualisasi dengan matplotlib dan seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.values, y=category_sales.index, palette='viridis')
plt.xlabel('Total Penjualan')
plt.ylabel('Kategori Pizza')
plt.title('Total Penjualan Berdasarkan Kategori Pizza')
plt.show()
```

```python
#Penjualan Berdasarkan Hari dalam Seminggu: Mengidentifikasi hari-hari tertentu dalam seminggu dengan penjualan tertinggi dan terendah.

# Pastikan kolom 'order_date' dalam format datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Mengonversi tanggal pesanan ke hari dalam seminggu
df['day_of_week'] = df['order_date'].dt.day_name()

# Menghitung total penjualan per hari dalam seminggu
weekly_sales = df.groupby('day_of_week')['total_price'].sum()

# Mengurutkan hari dalam seminggu
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sales = weekly_sales.reindex(days_order)

# Menampilkan hasil
print("Total Penjualan Berdasarkan Hari dalam Seminggu:")
print(weekly_sales)

# Visualisasi dengan matplotlib dan seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=weekly_sales.index, y=weekly_sales.values, palette='viridis')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Total Penjualan')
plt.title('Total Penjualan Berdasarkan Hari dalam Seminggu')
plt.xticks(rotation=45)
plt.show()

# Menghitung jumlah pesanan per hari dalam seminggu
weekly_order_count = df.groupby('day_of_week')['order_details_id'].count()

# Mengurutkan hari dalam seminggu
weekly_order_count = weekly_order_count.reindex(days_order)

# Menampilkan hasil
print("\nJumlah Pesanan Berdasarkan Hari dalam Seminggu:")
print(weekly_order_count)

# Visualisasi dengan matplotlib dan seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=weekly_order_count.index, y=weekly_order_count.values, palette='viridis')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Pesanan')
plt.title('Jumlah Pesanan Berdasarkan Hari dalam Seminggu')
plt.xticks(rotation=45)
plt.show()
```

```python
#Popularitas Bahan: Menghitung seberapa sering bahan tertentu muncul dalam pesanan untuk mengetahui bahan yang paling populer.

 #Menggabungkan semua bahan dalam satu list
all_ingredients = df['pizza_ingredients'].str.split(', ').sum()

# Menghitung frekuensi setiap bahan
ingredient_counts = Counter(all_ingredients)

# Mengonversi hasil ke DataFrame
ingredient_df = pd.DataFrame.from_dict(ingredient_counts, orient='index', columns=['count']).sort_values(by='count', ascending=False)

# Menampilkan 10 bahan terpopuler
print("Bahan Terpopuler:")
print(ingredient_df.head(10))

# Visualisasi dengan matplotlib dan seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x=ingredient_df.head(10).count, y=ingredient_df.head(10).index, palette='viridis')
plt.xlabel('Jumlah Penggunaan')
plt.ylabel('Bahan')
plt.title('10 Bahan Terpopuler dalam Pesanan')
plt.show()
```

```python
#Preferensi Bahan Berdasarkan Kategori: Melihat bahan yang paling sering digunakan dalam setiap kategori pizza.

# Fungsi untuk menghitung frekuensi bahan dalam setiap kategori
def count_ingredients_by_category(df):
    # Mengelompokkan data berdasarkan kategori pizza
    category_ingredient_counts = {}
    
    for category, group in df.groupby('pizza_category'):
        all_ingredients = group['pizza_ingredients'].str.split(', ').sum()
        ingredient_counts = Counter(all_ingredients)
        category_ingredient_counts[category] = ingredient_counts
    
    return category_ingredient_counts

# Menghitung frekuensi bahan dalam setiap kategori
category_ingredient_counts = count_ingredients_by_category(df)

# Menampilkan hasil untuk setiap kategori
for category, counts in category_ingredient_counts.items():
    print(f"\nBahan Terpopuler dalam Kategori {category}:")
    print(pd.DataFrame.from_dict(counts, orient='index', columns=['count']).sort_values(by='count', ascending=False).head(10))

# Visualisasi bahan terpopuler untuk setiap kategori
for category, counts in category_ingredient_counts.items():
    ingredient_df = pd.DataFrame.from_dict(counts, orient='index', columns=['count']).sort_values(by='count', ascending=False).head(10)
    
    plt.figure(figsize=(12, 8))
    sns.barplot(x=ingredient_df['count'], y=ingredient_df.index, palette='viridis')
    plt.xlabel('Jumlah Penggunaan')
    plt.ylabel('Bahan')
    plt.title(f'10 Bahan Terpopuler dalam Kategori {category}')
    plt.show()
```

```python
# Memeriksa format dari kolom 'order_time'
print(df['order_time'].unique())  # Melihat nilai unik di kolom order_time

# Mengonversi kolom 'order_time' ke format datetime
# Jika formatnya hanya jam, kita bisa menggunakan format '%H' atau '%H:%M'
# Sesuaikan format sesuai dengan data yang ada
try:
    df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M').dt.hour
except ValueError:
    df['order_time'] = pd.to_datetime(df['order_time'], format='%H').dt.hour

# Menghitung jumlah pesanan per jam
hourly_order_count = df.groupby('order_time')['order_details_id'].count().reset_index()

# Menampilkan hasil
print("Jumlah Pesanan per Jam:")
print(hourly_order_count)

# Visualisasi dengan matplotlib dan seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=hourly_order_count['order_time'], y=hourly_order_count['order_details_id'], palette='viridis')
plt.xlabel('Jam')
plt.ylabel('Jumlah Pesanan')
plt.title('Jumlah Pesanan Berdasarkan Waktu')
plt.xticks(rotation=90)
plt.show()

# Segmentasi pelanggan berdasarkan kebiasaan waktu
def segment_customers_by_order_time(df):
    df['order_time_segment'] = pd.cut(df['order_time'], bins=[0, 6, 12, 18, 24],
                                       labels=['Dini Hari', 'Pagi', 'Sore', 'Malam'])
    return df.groupby('order_time_segment')['order_details_id'].count().reset_index()

# Mengelompokkan pelanggan berdasarkan waktu pesanan
customer_segments = segment_customers_by_order_time(df)

# Menampilkan hasil segmentasi pelanggan
print("\nSegmentasi Pelanggan Berdasarkan Waktu Pesanan:")
print(customer_segments)

# Visualisasi segmentasi pelanggan
plt.figure(figsize=(10, 6))
sns.barplot(x=customer_segments['order_time_segment'], y=customer_segments['order_details_id'], palette='viridis')
plt.xlabel('Segmen Waktu Pesanan')
plt.ylabel('Jumlah Pesanan')
plt.title('Segmentasi Pelanggan Berdasarkan Waktu Pesanan')
plt.xticks(rotation=45)
plt.show()
```
